# ▶Part1. 토큰화
- 문장에서 토큰으로 나눌 구분자를 이용하여 단어 분리

->결과 [['파이썬', '차트', '파이썬', '머신러닝'], ['차트', '파이썬', 'R', '차트'], ['R', '분석', '시각화']]



```python
# 코드의 편의성을 위해서 txt자료는 빈칸단위로 분리할수 있는 자료준비
import numpy as np
txt=['파이썬 차트 파이썬 머신러닝',
      '차트 파이썬 R 차트',
       'R 분석 시각화'
      ]

#외부데이터 불러오기
f=open('샘플텍스트.txt','r',encoding='utf-8')
tmp=f.read()
tmp=tmp[1:len(tmp)-1]
txt=tmp.split('\n')
txt



  


```

```python
############################ 
##  [방법1] 빈칸단위로 나누기
###############################
txtList=[]
for i in txt:
  txtList.append(i.split(' '))
txtList
```

```python
############################ 
##  [방법2] 데이터 나누기 컴프리헨션
###############################

txtList=[i.split(' ') for i in txt]
txtList
```

# ▶Part2. 토큰 데이터 셋 1차원으로 변형
- 중복단어를 제거하여 고유값으로 필드를 제작하기 전 1차원 리스트로 변경해야함.

```python
############################ 
##  [방법1]  2차원리스트를 1차원으로 변형
###############################
import numpy as np

dataList=[]

for i in txtList:
  for j in i:
    dataList.append(j)


dataList
```

```python
############################ 
##  [방법2] 2차원 리스트 1차원 결합 컴프리헨션
###############################

dataList=[j for i in txtList for j in i]
dataList
```

```python
############################ 
##  [방법3] 넘파이를 이용하여서 한개의 리스트로 제작가능
###############################
dataList=np.concatenate(np.array(txtList))
dataList
```

# ▶Part4. 단어별 Count(빈도수 제작)

```python
#######################################
### DTM 제작, count함수 사용
### [방법1] count함수 사용
######################################
import pandas as pd
dtm=[]
for a in uniqueList:
  row=[]
  for i in txtList:
    row.append(i.count(a))
  dtm.append(row)
df=pd.DataFrame(dtm)
df=df.T
df.columns=uniqueList
df

numdf=df.values
print(np.sum(numdf,axis=0))
print(np.sum(numdf,axis=1))

df
```

```python
######################################
### 컬럼자료의 출현횟수를 계산함.
### 직접작성
### [방법2] count함수도 직접 제작
###################################
import pandas as pd
result=[]

for col in uniqueList:   # 리스트의처음값 'R'로 보기
    list_=[]
    for row in txtList:      #1번행에서 확인
        cnt=0
        
        for item in row:
            if col==item:
                cnt=cnt+1
        # print('------------------------------')
        #print('검색단어: %s   검색된횟수: %d' %(col,cnt))
        
        list_.append(cnt)
    result.append(list_)

df = pd.DataFrame(result)
print(df)

print('--------------------------')
df=df.T
print(df)


df.columns = uniqueList
display(df)

print(np.sum(df.values, axis=1))
print(np.sum(df.values, axis=0))
```

# ▶Part4. TF-IDF제작-1

```python
########################################################
##### Part4의 방법2임
#########################################################

txt=['파이썬 차트 파이썬 머신러닝',
      '차트 파이썬 R 차트',
       'R 분석 시각화'
      ]

vocab = list(set(w for doc in  txt for w in doc.split()))
vocab.sort()
vocab


def tf(t, d):
    return d.count(t)

result = []
N=len(txt)

for i in txt: 
 
  result.append([])
  for j in vocab:
        result[-1].append(tf(j, i))
    
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
from math import log # IDF 계산을 위해
```

```python
N=len(tf_)
print(N)
re=[]

for row in tf_.values:
  line=[]
 
  for col in row:
    line.append(log(N/(col+1)))
  re.append(line)
pd.DataFrame(np.array(result)*np.array(re),columns=vocab)

```

# ▶ Part5. TF-IDF제작-2
 - https://wikidocs.net/31698

```python
def tf(t, d):
    return d.count(t)

def idf(t):
    
    df = 0
    for doc in docs:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t, d):
  return tf(t,d)* idf(t)


def split(docs):
  return list(set(w for doc in docs for w in doc.split()))
```

```python
###############################################################
### TF-IDF제작
################################################################

import pandas as pd # 데이터프레임 사용을 위해
from math import log # IDF 계산을 위해

docs = [
  '먹고 싶은 사과',
  '먹고 싶은 바나나',
  '길고 노란 바나나 바나나',
  '저는 과일이 좋아요'
] 
vocab = split(docs)
vocab.sort()


N=len(docs)

result = []
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

```python
###############################################################
### 사이킷런 모듈 TF-IDF 
################################################################

from sklearn.feature_extraction.text import TfidfVectorizer
corpus = [

  '먹고 싶은 사과',
  '먹고 싶은 바나나',
  '길고 노란 바나나 바나나',
  '저는 과일이 좋아요'
   
]

tfidfv = TfidfVectorizer().fit(corpus)

# 열이름 가져오기
features=tfidfv.get_feature_names()
print('\n▶열이름 출력'+'='*100)
print(features)


######################################################################################
### 사이킷런에서 제공하는 TF-IDF
###
### TF-IDF 계산
#####################################################################################
# 일반 배열 구조로 변환함.
# 이 변환자료와 위의 단어자료를 합쳐서 데이터 프레임으로 작업하면됨
print('\n▶TF-IDF 계산행렬'+'='*100)
print(tfidfv.transform(corpus).toarray())

## 
print('\n▶'+'='*100)

tmp=tfidfv.transform(corpus).toarray()
import pandas as pd
# display(pd.DataFrame(tmp,columns=features))
display(pd.DataFrame(tmp,columns=features))
```

```python
############################################################################################
### TF-IDF 참고 
### analyzer 학습단위를 결정하는 파라미터
### analyzer = 'word'라고 설정시, 학습의 단위를 단어로 설정합니다. 
### (ex - home, go, my ...) analyzer = 'char'라고 설정시, 학습의 단위를 글자로 설정합니다.(ex - a, b, c, d ...)

### 단어: 단어를 오름차순으로 했을때의 순서번호
### 딥러닝에서도 단어분석시에 다음과 같이 딕셔너리의 형식으로 단어와 인덱스 번호를 갖게 됩니다.
### 인덱스 번호는 사용하는 패키지에 따라 방법으로 부여함. (오름차순 또는 출현횟수순등.)
#########################################################################################

vectorizer = TfidfVectorizer(analyzer = 'word')
vectorizer.fit(corpus)
print(sorted(vectorizer.vocabulary_.items()))

print('\n-----------------------------------------------------------')
print(vectorizer.transform(corpus))


```

```python
#####################################################
# 확률적 경사하강법을 이용한 문장 유사도 구하기
######################################################

from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [

  '전화 문의 불편 빠른 응답 전화 친절',
  '안내 불편 전화 무응답 전화 안내 불편',
  '배달비 비쌈 배달 느림 배달 불친절 안내 불편',
  '배달 용기 변경 요망 배달 속도 개선'
   
]

answer=[
        '전화관련과 안내관련 업무에 대해 다시한번 살펴보고 조치하겠습니다',
        '안내시스템 관련 부서로 연결하겠습니다.',
        '배달과 관련한 불편사항을 자세하게 말씀해주세요',
        '배달관련 부서에 전달하겠습니다'
]
       
tfidfv = TfidfVectorizer().fit(corpus)

X=tfidfv.fit_transform(corpus)
y=[1,2,3,4]

model=SGDClassifier(loss='perceptron')
model.fit(X,y)

newX=tfidfv.transform(['전화 연결 안됢'])
pred=model.predict(newX)
pred[0]
answer[pred[0]]
```
