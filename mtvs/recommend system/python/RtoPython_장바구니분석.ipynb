```python
import pandas as pd

data = pd.read_csv("C:\data\CUST_ORDER_OUTPUT2.utf8.csv")
```

```python
tmp = pd.DataFrame(data)
df = tmp[tmp['SEX'] != '*']
```

```python
df['Year'] = df['ORDER_DATE'].str.split('-').str[0]
df['Month'] = df['ORDER_DATE'].str.split('-').str[1]
df['Day'] = df['ORDER_DATE'].str.split('-').str[2]
```

```python
df
```

```python
def func(x) :
    if x == 'F' :
        return "female"
    elif x == 'M' :
        return "male"

df['gender'] = df['SEX'].apply(lambda x: func(x))
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

data_male = []
data_female = []

for i in range(1, 13):
    tmp = str(i).zfill(2)
    tmp2 = len(df[(df['Month']==tmp ) &  (df['gender']=='female')])
    data_female.append(tmp2)
    
    tmp3 = len(df[(df['Month']==tmp ) &  (df['gender']=='male')])
    data_male.append(tmp3)


plt.bar(range(len(data_female)), data_female , color = 'pink')
plt.bar(range(len(data_male)),data_male, bottom = data_female, color='skyblue')

plt.legend(['F','M'])
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.subplot(1,2,1)
plt.bar(range(len(data_female)), data_female)

plt.subplot(1,2,2)
plt.bar(range(len(data_male)), data_male)
plt.show()
```
