# <h> < FIFA 데이터 Columns 정보 > <h>
- id : 선수 고유의 아이디 </br>
- name : 이름</br>
- age : 나이</br>
- continent : 선수들의 국적이 포함되어 있는 대륙입니다</br>
- contract_until : 선수의 계약기간이 언제까지인지 나타내어 줍니다</br>
- position : 선수가 선호하는 포지션입니다. ex) 공격수, 수비수 등</br>
- prefer_foot : 선수가 선호하는 발입니다. ex) 오른발</br>
- reputation : 선수가 유명한 정도입니다. ex) 높은 수치일 수록 유명한 선수</br>
- stat_overall : 선수의 현재 능력치 입니다.</br>
- stat_potential : 선수가 경험 및 노력을 통해 발전할 수 있는 정도입니다.</br>
- stat_skill_moves : 선수의 개인기 능력치 입니다.</br>
- value : FIFA가 선정한 선수의 이적 시장 가격 (단위 : 유로) 입니다</br>

```python
#패키지 heatmap
# import folium
# from folium import plugins


# 세계지도 생성, 세계적으로 찍을 예정이기 때문에 따로 위도 경도는 지정해주지 않음
map = folium.Map(zoom_start= 30)


#train 데이터 불러 오는 함수,
train_df = pd.read_csv('FIFA_train.csv')
# test_df = pd.read_csv('.\\data\\FIFA_test.csv')

#필요한 자료만 불러옴, 대륙 정보, 대륙의 가중치 정보,
map_df = train_df[['continent']]
tmp  = map_df['continent'].value_counts()
tmp_df = pd.DataFrame(tmp)


#나라 별로 map 함수를 통해위도 값을 넣어주기 위한 함수
def converting(continent):
    if continent == 'south america':
        return -12.785707
    elif continent == 'europe':
        return 47.710319
    elif continent == 'africa':
        return 9.907480
    elif continent == 'oceania':
        return -27.270950
    else:
        return 37.5666805

#나라별로 map 함수를 통해 경도 값을 넣어주기 위한 함수
def convert2(continent):
    if continent == 'south america':
        return -57.647070
    elif continent == 'europe':
        return 6.986008
    elif continent == 'africa':
        return 20.315757
    elif continent == 'oceania':
        return 136.331382
    else:
        return 126.9784147

#맵 함수를 통해 위도와 경도를 데이터프레임에 추가해줌,
tmp_df['latitude'] = tmp_df['continent'].index.map(lambda x : converting(x))
tmp_df['longitude'] = tmp_df['continent'].index.map(lambda x : convert2(x))
map_df=tmp_df.copy()

#folium 패키지를 활용하여 위도 경도 가중치를 입력, heat map을 그림
map.add_child(plugins.HeatMap(zip(map_df['latitude'],map_df['longitude'],map_df['continent']), radius=20))
```

# Step 4 : 데이터 준비

### 왜도(Skewness): 분포의 비대칭도
- 정규분포 = 왜도 0
- 왼쪽으로 치우침 = 왜도 > 0
- 오른쪽으로 치우침 = 왜도 < 0

- 앞선 데이터 시각화를 통해 y변수인 'value'가 왼쪽으로 치우쳐져 있는 것을 확인하였다.
- 이를 해결하기 위해 3가지의 방법을 사용하였다.

방법1 ) 로그 변환

방법2 ) 루트 변환

방법3 ) boxcox 변환

- 방법 2) MinMaxScale
- 방법 3) StandardScale

```python
#아무작업 하지 않은 age변수에 대한 고민
# 값 유지
x_data['age']
#왼쪽 치우침, 왜도 > 로그 변환
x_data['age_log'] = np.log1p(x_data['age'])
#min-max scaling
x_data['age_minmax'] = (x_data['age']-x_data['age'].min())/(x_data['age'].max()-x_data['age'].min())
#standard scaling
x_data['age_std'] = (x_data['age']-x_data['age'].mean())/x_data['age'].std()
```

### linear regression

```python
#데이터셋별로 모델링 진행
for i, a in enumerate(x_dataset):
    for j, b in enumerate(y_dataset):
        X_train, X_test, y_train, y_test = train_test_split(a, b, test_size=0.3, random_state=9)

        reg = LinearRegression()
        reg.fit(X_train, y_train)

        pred_tr = reg.predict(X_train)
        pred_test = reg.predict(X_test)
        rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))
        rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

        x_name.append(x_dataset_name[i])
        y_name.append(y_dataset_name[j])
        tr_rmse.append(rmse_tr)
        test_rmse.append(rmse_test)
        train_score.append(reg.score(X_train,y_train))
        test_score.append(reg.score(X_test, y_test))
```

### xgboost tuning

- 다중공선성이 존재하지 않는 것을 확인하였으므로, parameter를 조정하는 것으로 score값을 올려보았다.

```python
x_name = []
y_name = []
tr_rmse = []
test_rmse = []
train_score=[]
test_score = []
c_name=[]
d_name=[]
e_name=[]
```

### Boost Parameter
- Learning rate(일반적으로 0.01 - 0.2)
- max_depth: Tree 깊이 수
- gamma: split 하기 위한 최소의 loss 감소 정의
- subsample: 데이터 중 샘플링(0.5 - 1)
- colsample_bytree: column 중 sampling(0.5 - 1)


```python
esti = [100, 300, 500, 700, 900]
lr = [0.01, 0.03, 0.05, 0.07, 0.09, 0.12, 0.15, 0.18, 0.2]
m_th = [2,3,4,5,6,7]
sub_samp=[0.5, 0.6, 0.7, 0.8]
cols_by=[0.5, 0.7, 0.9]

for a in esti:
    for b in lr:
        for c in m_th:
            for d in sub_samp:
                for e in cols_by:

                    X_train, X_test, y_train, y_test = train_test_split(x_data_basic, 
                                                                        y_data_log, 
                                                                        test_size=0.3, 
                                                                        random_state=9)

                    xgb = xgboost.XGBRegressor(n_estimators=a,
                                               learning_rate=b,
                                               gamma=0,
                                               subsample=d,
                                               colsample_bytree=e,
                                               max_depth=c)
                    xgb.fit(X_train, y_train)

                    pred_tr = xgb.predict(X_train)
                    pred_test = xgb.predict(X_test)
                    rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))
                    rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

                    x_name.append(a)
                    y_name.append(b)
                    c_name.append(c)
                    d_name.append(d)
                    e_name.append(e)
                    tr_rmse.append(rmse_tr)
                    test_rmse.append(rmse_test)
                    train_score.append(xgb.score(X_train,y_train))
                    test_score.append(xgb.score(X_test, y_test))
```

```python
df_xgboost = pd.DataFrame({'estimator' : x_name,
                  'learningrate' : y_name,
                  'max_depth' : c_name,
                  'sub_sample' : d_name,
                  'colsample_bytree' : e_name,
                  'train_rmse' : tr_rmse,
                  'test_rmse' : test_rmse,
                  'train_score' : train_score,
                  'test_score' : test_score})
```

- train score를 내림차순으로 살펴보면 parameter들이 높은 값들이다.
- test score를 기준으로 내림차순하여 살펴보면, max_depth=3과 learning rate=0.03이 눈에 띈다. 
- 무조건 높은 parameter를 찾기보다 적정한 parameter를 찾는 것이 좋은 모델링이라는 것을 어느정도 알 수 있는 대목이다.

- 찾은 parameter 모델로 제출한 모델의 점수는 909,143점으로 전의 모델보다 점수가 줄어든 것을 알 수 있다.

# 프로젝트를 통해 느낀점

 박이정 : 실제 축구 선수 데이터이기 때문에 이상치에 대한 의심은 하였지만, 제외하지는 않았습니다. 
         이번 데이터의 전처리 중 핵심은 왜도 처리 방식을 확실히 알 수 있어서 좋았고, 수업 시간에 배운 내용을 그대로 확인 할 수 있어서 좋았습니다. xgboost가 가장 성능이 좋다는 것을 확인하였으나, 
         계산량이 많았고, 가장 효율적인(결과, 성능, 소요시간 등) 알고리즘으로는 random forest라고 생각되어 현업에 적용한다면 random forest로 결정하지 않았을까..하는 생각을 했습니다. 
         linear regression의 경우, y변수 스케일링 여부에 따라 score값이 큰 폭으로 향상되므로 또 한번 전처리는 중요하다고 생각하였습니다. 
         모델 튜닝의 경우, parameter별 변화를 확인하기 위해 시도해본 것으로 무조건 높거나, 세세한 parameter보다 효율적이고 적정한 parameter를 찾는 것이 중요하다고 생각되었고, 
         개인적으로 크롤링을 하여 좀 더 정보를 붙여서 자세히 분석하고자 시도는 하였으나, 완성 시키지 못 한 점이 아쉽습니다. 
         첫 프로젝트를 함께한 조원분들이 다 열정적으로 임해주셔서 더 많이 배우는 시간이였습닌다! :) 

조우현 : AI과정을 시작하고 정신없이 배웠던 과정을 정리 해볼 수 있는 프로젝트라 뜻깊었다고 생각합니다. 
        작은 데이터에서 다양한 인사이트를 도출하고 다양한 방법으로 시각화를 진행하였습니다. 
        또한 다양한 모델을 학습시켜 보고 성능을 비교해보면서 짧지만 다양한 경험을 얻을수 있었습니다. 
        또한 팀원들의 작업 방식에서 전처리나, 모델학습 부분에서 많은 것을 배울수 있는 프로젝트 였습니다.

심현수 : 처음에는 시각화쪽에 큰 관심이 없었지만 팀원분들이 데이터 시각화를 통한 방법으로 인사이트를 실용적으로 
         도출해 내는 것을 보고 시각화의 중요성을 알 수 있던 계기가 되었습니다.
         linear regression, Random Forest, Xgboost, PCA, OLS등의 모델을 돌려보고 데이콘 사이트에서 
         직접적으로 오차율을 줄여보기 위해 노력했지만 아직 오차율을 크게 잡지는 못해 더욱 공부해야 겠다는 생각도 들었고 
         예측모델들의 결과를  test score순으로 정리하여 어떤 모델이 가장 실용적인지 직접 비교하는 방법을 생각한 팀원을 보고 
         좀 더 다양한 방법으로 데이터에 접근해봐야 겠단 생각도 들었습니다.
         이번 팀 프로젝트로 부족한점을 다시금 깨닫고 많은 성장을 이룰수 있게 도움을 주신 팀원분들께 다시금 감사드립니다.
