```python
# def fit_predict(model, X_train, y_train, X_test, y_test):
#     model = model()
#     model.fit(X_train, y_train)
#     y_pred = model.predict(X_test,y_test)
#     score = model.score(X_test, y_test)
    
#     return score, y_test, score
    
def regression_scores(y_true, y_pred) :

  import numpy as np
  from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error

  MSE = mean_squared_error(y_true, y_pred)
  RMSE = np.sqrt(mean_squared_error(y_true,y_pred))
  MAE = np.mean( np.abs((y_true - y_pred) / y_true) )
  NMAE =mean_absolute_error(y_true, y_pred)/ np.mean( np.abs(y_true) )
  MAPE = np.mean( np.abs((y_true - y_pred) / y_true) ) *100
  R2 = r2_score(y_true, y_pred)
  return {'MSE' : np.round(MSE, 3), 
          'RMSE' : np.round(RMSE, 3),
          'MAE' : np.round(MAE, 3),
          'NMAE' : np.round(NMAE, 3),
          'MAPE' : np.round(MAPE, 3),
          'R2' : np.round(R2, 3)}
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
score = lr.score(X_test, y_test)
print(f"r2_score : {score}")

regression_scores(y_test, pred)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
score = rf.score(X_test, y_test)
print(f"r2_score : {score}")

regression_scores(y_test, pred)
```

```python
xr = XGBRegressor()
xr.fit(X_train, y_train)
pred = xr.predict(X_test)
score = xr.score(X_test, y_test)
print(f"r2_score : {score}")

regression_scores(y_test, pred)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(

X, y)
```
